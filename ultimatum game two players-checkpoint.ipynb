```python
def game():
    
    random_number_a = random.random()
    random_number_b = random.random()
    a_done = False
    b_done = False
    a_sum = 0
    b_sum = 0
    for i in range(L+1):
        if a_done == False:
            if random_number_a <= a_sum + p_a[i]:
                a_offer = i/L
                a_done = True
            else:
                a_sum += p_a[i]
        if b_done == False:
            if random_number_b <= b_sum + p_b[i]:
                b_threshold = i/L
                b_done = True
            else:
                b_sum += p_b[i]
        if a_done == True and b_done == True:
            break
    #a_offer = 0.75
    #b_threshold = 0.75
    if a_offer == 0.0:
        c_decision = 0
    elif a_offer == 1.0:
        c_decision = 1
    elif a_offer >= b_threshold:
        c_decision = 1
    else:
        c_decision = 0
    a_vector.append(a_offer)
    b_vector.append(b_threshold)
    c_vector.append(c_decision)
    print("game outcome: ", a_offer, c_decision)
    return (1-a_offer)*c_decision
```

```python
def probabilityupdate(key, w_a, w_b, t=1):
    sum_a = 0
    sum_b = 0
    positive_components = 0
    to_add_to_positive_components = 0
    if key == 'softmax':
        for i in range(L+1):
            sum_a += math.exp(w_a[i]/t)
            sum_b += math.exp(w_b[i]/t)
        for i in range(L+1):
            p_a[i] = math.exp(w_a[i]/t)/sum_a
            p_b[i] = math.exp(w_b[i]/t)/sum_b
    else:
        for i in range(L+1):
            sum_a += w_a[i]
            sum_b += w_b[i]            
        for i in range(L+1):
            p_a[i] = w_a[i]/sum_a
            p_b[i] = w_b[i]/sum_b
```

```python
def phigradient(gamenumber,gradient_pa):
    result = np.zeros(L+1)
    if c_vector[gamenumber] == 0:
        for i in range(round(a_vector[gamenumber]*L) + 1,L+1):
            result += ((1-i/L)*(i/L-a_vector[gamenumber])/(1-a_vector[gamenumber]))*np.array(gradient_pa[i])
    else:
        #print("no. of terms in sum", round(a_vector[gamenumber]*L)+1)
        for i in range(round(a_vector[gamenumber]*L) + 1):
            #print(i)
            result += ((1-i/L)*(i/a_vector[gamenumber]))*np.array(gradient_pa[i])
            #print(result)
    #print(result)                 
    return result
```

```python
def expectation_of_next_agreement(w_a,w_b,t):
    Z = 0.0
    result = 0.0
    for i in range(L+1):
        for j in range(0,i+1):
            Z += math.exp((w_a[i] + w_b[j])/t)
    for i in range(L+1):
        prob = 0
        for j in range(0,i+1):
            prob += math.exp((w_a[i] + w_b[j])/t)
        result += (i/L)*prob
    return result/Z
```
