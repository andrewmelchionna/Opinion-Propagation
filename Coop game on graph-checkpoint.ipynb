```python
def weightupdate():
    for i in range(n):
        w[i][0] = coordinates[i][0]*reward[0]
        w[i][1] = coordinates[i][1]*reward[1]
```

```python
def probabilityupdate(function):
    result = [0]*n
    if function == 'squares':
        for i in range(n):
            result[i] = w[i][0]**2/(w[i][0]**2 + w[i][1]**2)
    elif function == 'softmax':
        for i in range(n):
            result[i] = math.exp(w[i][0])/(math.exp(w[i][0]) + math.exp(w[i][1]))
    elif function == 'polya':
        for i in range(n):
            result[i] = w[i][0]/(w[i][0] + w[i][1])
    return result
```

```python
def game_completegraph():
    i = random.randint(0, n-1)
    j = i
    while j == i:
        j = random.randint(0, n-1)
        
    qi = random.random()
    qj = random.random()
    if qi < p[i]:
        a = 0
    else:
        a = 1
    if qj < p[j]:
        b = 0
    else:
        b = 1
    if a == b:
        coordinates[i][a] += 1
        coordinates[j][a] += 1
```

```python
def game_alwaysagree(adj_matx, function):
    foundedge = False
    while foundedge == False:
        i = random.randint(0, n-1)
        j = random.randint(0, n-1)
        if adj_matx[i][j] == 1:
            foundedge = True
    if function == 'polya':
        p = (w[i][0] + w[j][0])/(w[i][0] + w[j][0] + w[i][1] + w[j][1])
    elif function == 'softmax':
        p = math.exp(w[i][0]+w[j][0])/(math.exp(w[i][0]+w[j][0]) + math.exp(w[i][1]+w[j][1]))
    q = random.random()
    #print("i,j ", i,j)
    if q < p:
        coordinates[i][0] += 1
        coordinates[j][0] += 1
        #print("coordinates immediately after game ", coordinates)
    else:
        coordinates[i][1] += 1
        coordinates[j][1] += 1
        #print(1)
```

```python
def summedmartingale(x,y):
    result = 0
    if x != 0:
        for i in range(1,x):
            result += 1/(2*i)
    if y!= 0:
        for i in range(1,y):
            result -= 1/i
    return result
        
```

```python
def martingaleplot(endingmartingale,endpoint,numberofbins):
    binwidth = 2*endpoint/(numberofbins-2)
    counts = np.zeros(numberofbins)
    for i in range(len(endingmartingale)):
        adjustedvalue = np.floor((endingmartingale[i] + endpoint)/binwidth)
        #print(adjustedvalue)
        if adjustedvalue < 0:
            counts[0] += 1
        elif adjustedvalue >= numberofbins - 2:
            counts[numberofbins-1] += 1
        else:
            counts[round(adjustedvalue)+1] += 1
    return counts
```
